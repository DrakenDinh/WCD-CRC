```python
import os
import pandas as pd

# Input folder path (containing .xlsx files)
input_folder = r'C:\your_path\LR_Train_107_Shiftx30x4'

# Output Excel file path (merged Excel file)
output_excel_file = r'C:\your_path\LR_Train_107_Shiftx30x4_MER.xlsx'

# Collect all .xlsx files in the folder (exclude the output merged file if it already exists)
xlsx_files = [os.path.join(input_folder, file) for file in os.listdir(input_folder)
              if file.endswith('.xlsx') and file != os.path.basename(output_excel_file)]

# Read the first Excel file and create the output Excel file
first_xlsx_df = pd.read_excel(xlsx_files[0])
writer = pd.ExcelWriter(output_excel_file, engine='xlsxwriter')
first_xlsx_df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.close()

# Append the remaining .xlsx files as separate sheets
with pd.ExcelWriter(output_excel_file, engine='openpyxl', mode='a') as writer:
    for xlsx_file in xlsx_files[1:]:
        sheet_name = os.path.splitext(os.path.basename(xlsx_file))[0]
        df = pd.read_excel(xlsx_file)
        # Limit sheet name length to a maximum of 31 characters (Excel limitation)
        sheet_name = sheet_name[:31]
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```

```python
import pandas as pd
import numpy as np

excel_file = pd.ExcelFile(
    r'C:\your_path\LR_Train_107_Shiftx30x4_MER.xlsx'
)

all_data = []

for sheet_name in excel_file.sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
        
    column_data = df.iloc[:, 1:4].values
    # 1:3 means starting from column index 1 and selecting columns 2 and 3
    # 1:4 means starting from column index 1 and selecting columns 2, 3, and 4
    all_data.append(column_data)

all_data_array = np.array(all_data)

np.save(
    r'C:\your_path\LR_Train_107_Shiftx30x4_MER.npy',
    all_data_array
)

print("Data saved to all_data.npy")
```

```python
import numpy as np

# Paths to two numpy files: low trước, High sau
file_path_0 = r'C:\your_path\LR_Train_107_Shiftx30x4_MER.npy'
file_path_1 = r'C:\your_path\HR_Train_107_Shiftx60x2_MER.npy'

# Load numpy arrays
data_0 = np.load(file_path_0)
data_1 = np.load(file_path_1)

# Assign labels to each data
labels_0 = np.zeros(data_0.shape[0], dtype=np.int32)  # Label as 0 - Low risk
labels_1 = np.ones(data_1.shape[0], dtype=np.int32)   # Label as 1 - High risk

# Concatenate data and labels
combined_data = np.concatenate((data_0, data_1), axis=0)
combined_labels = np.concatenate((labels_0, labels_1), axis=0)

# Save as a single numpy array
np.save(r'C:\your_path\Train_107_SPA_GlobalMM_3waves_Shift_LR30x4_HR60x2_3030data_Dataset.npy', combined_data)
np.save(r'C:\your_path\Train_107_SPA_GlobalMM_3waves_Shift_LR30x4_HR60x2_3030data_Labels.npy', combined_labels)
```
